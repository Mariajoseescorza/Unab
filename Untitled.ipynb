```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página de Yahoo Finance de GME
url = "https://finance.yahoo.com/quote/GME/financials"

# Realizar la solicitud HTTP
response = requests.get(url)

# Comprobar si la solicitud fue exitosa
if response.status_code != 200:
    print(f"Error al obtener la página: {response.status_code}")
else:
    # Parsear el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscar todas las tablas en la página
    tables = soup.find_all('table')

    # Verificar si se encuentran tablas en la página
    if len(tables) > 0:
        # Seleccionar la primera tabla que contiene los datos financieros
        table = tables[0]  # Asegúrate de que esta es la tabla correcta

        # Extraer todas las filas de la tabla
        rows = table.find_all('tr')

        # Lista para almacenar los datos de ingresos
        data = []

        # Iterar sobre las filas de la tabla
        for row in rows:
            cols = row.find_all('td')
            if len(cols) > 1:  # Ignorar filas vacías
                year = cols[0].text.strip()  # Año
                revenue = cols[1].text.strip()  # Ingresos
                data.append([year, revenue])

        # Convertir los datos extraídos en un DataFrame de pandas
        gme_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

        # Mostrar las últimas 5 filas utilizando tail()
        print(gme_revenue.tail())
    else:
        print("No se encontraron tablas en la página.")
```

```python
import yfinance as yf
import pandas as pd

# Obtenemos los datos financieros de GameStop (GME)
gme = yf.Ticker("GME")

# Extraemos el estado financiero
gme_financials = gme.financials

# Mostrar los ingresos (Revenue) del DataFrame de resultados
gme_revenue = gme_financials.loc['Total Revenue']

# Convertir a DataFrame y mostrar las últimas 5 filas
gme_revenue_df = pd.DataFrame(gme_revenue).transpose()

# Mostrar las últimas 5 filas usando tail()
print(gme_revenue_df.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Obtener los datos históricos de las acciones de Tesla (TSLA)
tesla = yf.Ticker("TSLA")

# Obtener los datos de cierre ajustado (adjusted close) para el gráfico
tesla_data = tesla.history(period="1y")  # Datos de 1 año

# Crear una función para graficar
def make_graph(data):
    plt.figure(figsize=(10, 6))  # Tamaño del gráfico
    plt.plot(data.index, data['Close'], label='Precio de cierre ajustado')
    plt.title('Acciones de Tesla (TSLA) - Precio de Cierre Ajustado')
    plt.xlabel('Fecha')
    plt.ylabel('Precio en USD')
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.show()

# Llamar a la función para representar el gráfico
make_graph(tesla_data)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Obtener los datos históricos de las acciones de GameStop (GME)
gamestop = yf.Ticker("GME")

# Obtener los datos de cierre ajustado (adjusted close) para el gráfico
gamestop_data = gamestop.history(period="1y")  # Datos de 1 año

# Crear una función para graficar
def make_graph(data):
    plt.figure(figsize=(10, 6))  # Tamaño del gráfico
    plt.plot(data.index, data['Close'], label='Precio de cierre ajustado', color='blue')
    plt.title('Acciones de GameStop (GME) - Precio de Cierre Ajustado')
    plt.xlabel('Fecha')
    plt.ylabel('Precio en USD')
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.show()

# Llamar a la función para representar el gráfico
make_graph(gamestop_data)

```
